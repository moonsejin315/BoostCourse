# 2.2 판다스 치트시트를 활용한 기초 익히기

**판다스**는 데이터 분석을 위한 파이썬 라이브러리입니다. 데이터 프레임을 사용해 다양한 데이터 조작과 분석을 수행할 수 있습니다.

### 주요 판다스 기능
- to_csv("파일명", index=False) : csv 파일로 저장하기
- read_csv("파일명") : csv 파일 불러오기
- Series : 1차원 벡터구조
- DataFrame : 2차원 행렬구조
- Subset : 일부 값만 불러오기
- Summarize Data : 데이터 요약하기
    - .value_counts()
- Reshaping : 데이터 재정의
    - .sort_value()
    - .drop()
-Group Data : 데이터 묶기
    - .groupby() 
    - .pivot_table
- [컬럼]
- .loc[행]
- .loc[행, 열]
- .loc[조건식, 열]

### 시각화_Plotting
- .plot() : 꺾은선 그래프
- .plot.bar() : 막대그래프
- .plot.density() : 밀도함수
